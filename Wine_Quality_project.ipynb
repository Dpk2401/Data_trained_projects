## Removing outliers

Z-Score Technique

```python
from scipy.stats import zscore
import numpy as np
z=np.abs(zscore(df))
z.shape
```

```python
threshold=3    #rule of thumb
print(np.where(z>3))
```

```python
len(np.where(z>3) [0])
```

```python
z.iloc[4]
```

```python
df_new=df[(z<3).all(axis=1)]
print("Old DataFrame", df.shape)
print("New DataFrame", df_new.shape)
print("total_drop_rows", df.shape[0] - df_new.shape[0])
```

## Transform Data to remove skewness

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
from sklearn.metrics import accuracy_score
```

## Splitting the DataFrame

```python
for i in range(0, 1500):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=i, test_size=0.20)
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    if round(accuracy_score(y_train,pred_train)*100,1) == round(accuracy_score(y_test,pred_test)*100,1):
        print("At random state", i, "The Model performs very well")
        print("At random_state:-", i)
        print("Training accuracy score is:-", accuracy_score(y_train,pred_train)*100)
        print("Training accuracy score is:-", accuracy_score(y_test, pred_test)*100, '\n')
```
